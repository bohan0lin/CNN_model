# cnn_util

```python
import numpy as np
import math
import pickle as pkl
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import matplotlib.pyplot as plt
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.metrics import accuracy_score
from torch.autograd import Variable

torch.manual_seed(0)

# data loader function, return four datasets: train_data_x, train_data_y, test_data_x, test_data_y
def load_dataset():
    train_data_x = pd.read_pickle("train_data_x.pkl")

    train_data_y = pd.read_pickle("train_data_y.pkl")

    test_data_x = pd.read_pickle("test_data_x.pkl")

    test_data_y = pd.read_pickle("test_data_y.pkl")

    return train_data_x, train_data_y, test_data_x, test_data_y

# CNN model
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        # The first convolutional layer
        self.conv1 = nn.Sequential(
            nn.Conv2d(
                in_channels = 3,
                out_channels = 6,
                kernel_size = 5,
                stride = 1
            ),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2)
        )
        # The second convolutional layer
        self.conv2 = nn.Sequential(
            nn.Conv2d(6, 12, 5, 1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2)
        )
        # Fully connected layer with activation function
        self.fc1 = nn.Linear(2028, 120)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(120, 64)
        self.fc3 = nn.Linear(64, 2)
        self.dropout = nn.Dropout(p = 0.2)
        self.sigmoid = torch.nn.Sigmoid()
    
    
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.dropout(x)
        
        x = x.view(x.size(0), -1)
        
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc3(x)
        x = self.sigmoid(x)

        return x
```
